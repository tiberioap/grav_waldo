### **WALDO: Waveform AnomaLy DetectOr**

```python
from waldo.waldo import Waldo
```

```python
waldo = Waldo(path_dset="dataset/dset_wfsHM", path_unet="dataset/unet", path_waldo="dataset/waldo", save_load=2)

print(waldo.__doc__)
```

### **Histogram setup:**

```python
ps = ["q < 5.0",
      "abs(chi11) <= 1.0", 
      "abs(chi12) <= 1.0", 
      "abs(chi13) <= 1.0", 
      "abs(chi21) <= 1.0", 
      "abs(chi22) <= 1.0", 
      "abs(chi23) <= 1.0",
      "e < 1.0"]

waldo.plot_histogram(l=4, m=4, quantile=0.99, bins=50, save=False, save_path="plots/", param_space=ps)
```

### **Plotting wfs with high mismatch:**

```python
waldo.plot_waveforms(x_limit=[-200, 100], save=False, save_path="plots/", wf_label="h_lm")
```
